## 1.1 Create a two-dimensional vector and plot it on a graph

```python
import numpy as np
import matplotlib.pyplot as plt

def Map(func, x):
    return list(map(func, x))


def step_lazy(func, iterable, args=[]):
    for item in iterable:
        yield func(item, *args)
            
def step(func, iterable, args=[], kwargs=[]):
    return list(step_lazy(func, iterable, args))

fig = plt.gcf()
v = np.array([1, 2]).T
plt.ylim(0,4)
plt.xlim(0,2)
plt.arrow(*np.repeat(0,2),*v, head_width=.05)
```

```python
print("CD\n", np.matmul(C, D)) # 2(C) rows by 3(D) columns
print("AE\n", np.matmul(A, E)) # 3(A) rows by 2(E) columns
print("BA\n", np.matmul(B, A)) # 1(B) rows by 2(A) columns
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
H = np.array([[12, 11],
              [7, 10]])
J = np.array([[0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]])
print('|H|', np.linalg.det(H))
print('|J|', np.linalg.det(J))
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```python
print('inverse of H', np.linalg.inv(H))
print('inverse of J', np.linalg.inv(J))
```
